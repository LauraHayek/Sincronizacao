```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi, 100) #x é um vetor de 100 pontos de 0 a 2pi
y = np.sin(x) #y calcula o seno para cada ponto de x e retorna um vetor

plt.plot(x, y) #grafico contínuo 
plt.scatter(x, y) #pontos no gráfico
plt.show()
```

```python
#exemplo com pontos definidos

x = [0.8, 2.2, 2.8, 3.8, 5.2, 6.3, 7.3, 7.9, 9.4, 10.0]
y = [4.0, 6.0, 7.5, 12, 15, 17.5, 22, 23, 27, 30]

plt.scatter(x, y, color = 'black')
plt.xlabel('t (s)')
plt.ylabel('S (cm)')
plt.grid()
plt.show()
```

```python
#exemplo usando numpy

x = np.array([0.8, 2.2, 2.8, 3.8, 5.2, 6.3, 7.3, 7.9, 9.4, 10.0])
y = np.array([4.0, 6.0, 7.5, 12, 15, 17.5, 22, 23, 27, 30])

c = np.polyfit(x, y, 1) #fit de um polinomio de grau 1 y = c0*x^1 + c1
yf = c[0]*x + c[1]

plt.scatter(x, y, color = 'black')
plt.plot(x, yf, color = 'red')
plt.xlabel('t (s)')
plt.ylabel('S (cm)')
plt.grid()
plt.show()
```

```python
#adicionando barras de erro

x = np.array([0.8, 2.2, 2.8, 3.8, 5.2, 6.3, 7.3, 7.9, 9.4, 10.0])
y = np.array([4.0, 6.0, 7.5, 12, 15, 17.5, 22, 23, 27, 30])
yerr = np.array([2, 3, 3, 2, 1, 4, 2, 3, 2, 3]) #yerr adiciona barras de erro no eixo y, xerr "()" no eixo x

c = np.polyfit(x, y, 1) #fit de um polinomio de grau 1 y = c0*x^1 + c1
yf = c[0]*x + c[1]

plt.errorbar(x, y, yerr = yerr, fmt = 'ob')
plt.scatter(x, y, color = 'black')
plt.plot(x, yf, color = 'red')
plt.xlabel('t (s)')
plt.ylabel('S (cm)')
plt.grid()
plt.show()

#O parâmetro fmt na função plt.errorbar() define o estilo dos marcadores usados para representar os pontos de dados no gráfico.

#fmt	Símbolo	Descrição
#'o'	●	    Círculo
#'s'	■	    Quadrado
#'^'	▲	    Triângulo para cima
#'v'	▼	    Triângulo para baixo
#'x'	✕	    X
#'D'	◆	    Diamante

#Exemplos de fmt personalizados
#Quadrados azuis (s e b): plt.errorbar(x, y, yerr=yerr, fmt='sb')  # s = quadrado, b = azul
#Triângulos vermelhos (^ e r): plt.errorbar(x, y, yerr=yerr, fmt='^r')  # ^ = triângulo, r = vermelho
#Pontos pretos sem linhas (. e k): plt.errorbar(x, y, yerr=yerr, fmt='.k')  # . = ponto, k = preto

#Se você quiser apenas exibir as barras de erro sem um marcador, pode definir fmt=''.
```

```python
#usando Latex na legenda

# Desativa o uso do LaTeX
#plt.rcParams['text.usetex'] = False
plt.rcParams.update({'font.size': 12})

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = r'$\sin(\theta)$')
plt.plot(x, y2, label = r'$\cos(\theta)$')
plt.legend(fontsize = 14)
plt.xlabel(r'$\theta$ [rad]')
plt.ylabel(r'Funções trigonométricas')
plt.tight_layout()
plt.show()

#O r antes de uma string em Python significa que você está criando uma raw string (string bruta). 
#No contexto do seu código, o r serve para garantir que os caracteres de escape dentro da string, como \, 
#sejam tratados literalmente, e não como caracteres especiais.
#O r assegura que a string será passada para o Matplotlib exatamente como está, incluindo o símbolo de barra invertida 
#(\) usado pelo LaTeX para denotar operações matemáticas.
```

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.rcParams['text.usetex'] = False

plt.figure(figsize=(6, 5))
plt.plot(x, y1, label=R'$\sin\theta$')
plt.plot(x, y2, label=R'$\cos\theta$', ls='--')
plt.plot(x, y3, label=R'$\tan\theta$', ls='-.')
plt.legend(fontsize=15, loc=(0, 1.05), ncol=3) #tamanho, distancia do grafico, tamanho da legenda
plt.xlabel(R'$\theta$ [rad]')
plt.ylabel(R'Trigonometric functions')

plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], ["0", R"$\dfrac{\pi}{2}$", R"$\pi$", R"$\dfrac{3\pi}{2}$", R"$2\pi$"])
#o primeiro argumento determina a posição, o segundo argumento determina a "legenda"

plt.xlim(0, 2*np.pi)
plt.ylim(-2, 2)
plt.grid()
plt.tight_layout()
plt.savefig('ex2-step4.png', dpi=300)
plt.show()

#O comando plt.tight_layout() em Matplotlib ajusta automaticamente os parâmetros de layout da figura para garantir que o conteúdo 
#(como títulos, rótulos dos eixos, legendas, etc.) não se sobreponha e que os elementos do gráfico fiquem bem distribuídos.
#Por exemplo, se você estiver usando rótulos grandes ou muitos subgráficos, plt.tight_layout() ajuda a ajustar o espaçamento entre 
#os elementos para que tudo seja exibido corretamente, sem cortar nada.
#Isso pode ser muito útil quando há muitos elementos na sua figura, como títulos ou rótulos longos, que de outra forma poderiam ser truncados ou sobrepostos.
```

```python
#paineis (subgraficos) e inserções

plt.rcParams.update({'font.size': 20})
plt.rcParams['text.usetex'] = False

x = np.linspace(0, 5, 2000)
y = np.sin(200*x) * np.exp(-x)

plt.figure()
plt.plot(x, y)
plt.xlabel(R'$t$')
plt.ylabel(R'$y(t)$')

ax = plt.gca() 
#ax = plt.gca() > gca() significa "get current axis" (obter o eixo atual). Esse comando retorna o objeto do eixo atual, 
#ou seja, o gráfico onde você está desenhando. Caso você não tenha um gráfico com eixos criados, o gca() cria um novo eixo.
#Isso é útil para modificar as propriedades do gráfico ou adicionar elementos ao gráfico que você já criou 
#(como títulos, labels ou linhas adicionais).

ins = ax.inset_axes([0.6, 0.6, 0.37, 0.37])
#adiciona um gráfico dentro de outro gráfico, ou seja, cria uma área (inserção) dentro do gráfico principal.
#O primeiro e o segundo valores (0.6, 0.6) são as coordenadas do canto inferior esquerdo da área de inserção, 
#em relação ao gráfico original (valores entre 0 e 1, onde 0 é o mínimo e 1 é o máximo de cada eixo).
#Os dois últimos valores (0.37, 0.37) definem a largura e altura da área de inserção.
#O ins se refere ao objeto do eixo inserido, que você pode usar para customizar esse subgráfico, como adicionar títulos, linhas, etc.

ins.plot(x, y)
ins.set_xlim(1.5, 1.7)
ins.set_ylim(-0.3, 0.3)
ins.set_xticklabels('') #nao mostrar os numeros nos eixos
ins.set_yticklabels('')

ax.indicate_inset_zoom(ins)

plt.tight_layout()
plt.show()
```

```python

plt.rcParams.update({'font.size': 20})
plt.rcParams['text.usetex'] = False

x = np.linspace(-10, 10, 100)
y1 = np.sin(x)/x
y2 = np.exp(-x**2)
y3 = np.tanh(x)
y4 = x**2

plt.figure(figsize=(8,5))

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.xlabel(R'$x$')
plt.ylabel(R'$\sin(x)/x$')
plt.text(0.05, 0.75, R'(a)', transform=plt.gca().transAxes)

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.xlabel(R'$x$')
plt.ylabel(R'$g(x)$')
plt.text(0.05, 0.75, R'(b)', transform=plt.gca().transAxes)

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.xlabel(R'$x$')
plt.ylabel(R'$\tanh(x)/x$')
plt.text(0.05, 0.75, R'(c)', transform=plt.gca().transAxes)

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.xlabel(R'$x$')
plt.ylabel(R'$x^2$')
plt.text(0.15, 0.75, R'(d)', transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()
```

```python
#plt.text(x, y, texto, transform=...) > O comando plt.text() insere um texto em uma posição específica dentro do gráfico.
#x e y: Coordenadas onde o texto será colocado.
#texto: O conteúdo do texto a ser exibido.
#transform=plt.gca().transAxes: Define o sistema de coordenadas no qual as posições x e y serão interpretadas.

#Analisando os parâmetros
#exemplo: 0.05, 0.75
#Essas são as coordenadas onde o texto será colocado. Mas como estamos usando transAxes, os valores vão de 0 a 1 e são relativos 
#ao tamanho do eixo, não às coordenadas dos dados do gráfico.
#0.05 → 5% da largura do eixo (perto do canto esquerdo).
#0.75 → 75% da altura do eixo (mais para o topo).

#R'(a)':
#O prefixo R'...' indica uma string "raw" (crua), ou seja, caracteres de escape (\n, \t, etc.) não são interpretados.
#No caso de '(a)', não faz diferença aqui, mas pode ser útil para fórmulas matemáticas mais complexas.
#O texto inserido será (a), que pode ser usado, por exemplo, para legendar figuras.

#transform=plt.gca().transAxes:
#isso faz com que as coordenadas (0.05, 0.75) sejam interpretadas em relação ao tamanho do eixo do gráfico, em vez de estarem 
#na escala dos dados (data coordinates).
#plt.gca() significa "get current axis" (pegar os eixos atuais).
#transAxes transforma as coordenadas para que 0,0 seja o canto inferior esquerdo do eixo e 1,1 seja o canto superior direito.

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)

# Adicionando um texto no gráfico
plt.text(0.05, 0.75, R'(a)', transform=plt.gca().transAxes, fontsize=14, color='red')

plt.show()
```

```python
#O comando plt.subplot(2, 2, 1) cria subgráficos (ou seja, múltiplos gráficos dentro da mesma figura). Vamos entender cada parte:
#Estrutura do plt.subplot(nrows, ncols, index)
#plt.subplot(nrows, ncols, index)
#nrows → Número de linhas da grade de subgráficos.
#ncols → Número de colunas da grade de subgráficos.
#index → Qual subgráfico está sendo ativado (numerado da esquerda para a direita, de cima para baixo).

#O que plt.subplot(2, 2, 1) faz?
#2, 2 → Cria uma grade 2×2 de subgráficos (ou seja, 4 gráficos no total).
#1 → Ativa o primeiro subgráfico (no canto superior esquerdo).

#A numeração dos subgráficos segue esta ordem:
#(2,2,1)  →  (2,2,2)
#(2,2,3)  →  (2,2,4)

#┌───────┬───────┐
#│   1   │   2   │
#├───────┼───────┤
#│   3   │   4   │
#└───────┴───────┘

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x)
y4 = np.log(x + 1)

plt.figure(figsize=(8, 6))  # Tamanho da figura

plt.subplot(2, 2, 1)  # Primeiro subgráfico
plt.plot(x, y1, 'r')
plt.title("Seno")

plt.subplot(2, 2, 2)  # Segundo subgráfico
plt.plot(x, y2, 'g')
plt.title("Cosseno")

plt.subplot(2, 2, 3)  # Terceiro subgráfico
plt.plot(x, y3, 'b')
plt.title("Exponencial decrescente")

plt.subplot(2, 2, 4)  # Quarto subgráfico
plt.plot(x, y4, 'm')
plt.title("Logaritmo")

plt.tight_layout()  # Ajusta espaçamento
plt.show()
```
