```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
N = 100
K = 1
t_max = 50
t = np.linspace(0, t_max, 1000)
teta_0 = np.random.uniform(0, 2*np.pi, N)

def kuramoto(teta, t, K, N):
    dteta = np.zeros(N)
    for i in range(N):
        soma_sen = np.sum([np.sin(teta[j] - teta[i]) for j in range(N)])
        dteta[i] = K/N * soma_sen
    return dteta

sol = odeint(kuramoto, teta_0, t, args = (K, N))
sol_2 = sol%(2*np.pi)

for i in range(N):
    plt.plot(t, sol_2[:, i])

plt.xlabel('Time (t)')
plt.ylabel('Phase ()')
plt.title('Kuramoto Model, 'f' N = {N}')
plt.show()
```
