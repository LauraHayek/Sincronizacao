```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

plt.rcParams['figure.dpi'] = 300
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
```

```python
valores_N = [20, 50, 100]
valores_p = [.2, .5, .8]

for N in valores_N:
    for p in valores_p:
        dist = np.random.binomial(N, p, 1000)

        plt.hist(dist, bins=np.arange(0, N+2)-0.5, density=True, alpha=0.6, color='b', edgecolor='black')

        x = np.arange(0, N+1)
        curva_teorica = binom.pmf(x, N, p)

        plt.plot(x, curva_teorica, 'r-', label='Curva Teórica')

        plt.xlabel('N')
        plt.ylabel('Frequência')
        plt.title('Distribuição Binomial para 'f'N = {N} e 'f'p = {p}')
        plt.legend()

        plt.show()
```
